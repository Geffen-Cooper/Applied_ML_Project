```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
```

```python
longitude_train = df_train[df_train['City'] == 'Atlanta']['Longitude']
latitude_train = df_train[df_train['City'] == 'Atlanta']['Latitude']

longitude_test = df_test[df_test['City'] == 'Atlanta']['Longitude']
latitude_test = df_test[df_test['City'] == 'Atlanta']['Latitude']
```

```python
fig,ax = plt.subplots(1,1,dpi=200)
ax.scatter(longitude_train,latitude_train,s=3,c='k',label='train intersection')
ax.scatter(longitude_test,latitude_test,s=1,c='r',label='test intersection')
ax.set_xlabel('Longitude')
ax.set_xlabel('Latitude')
ax.set_title('Atlanta')
ax.legend()
plt.savefig("intersections.png")
```
